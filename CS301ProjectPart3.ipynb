```python
import numpy as np 
import matplotlib.pyplot as plt 
  
  
#Creating dataset 
States = ['A-E States', 'F-J States', 'K-O States', 'P-T States','U-Z States']

Pub_InStateData=[PC_IST_AvgA_E, PC_IST_AvgF_J, PC_IST_AvgK_O, PC_IST_AvgP_T, PC_IST_AvgU_Z,]
     
# Creating explode data 
BlowOut = (0.1, 0.0, 0.2, 0.1, 0.2) 
  
# Creating color parameters 
colors = ( "cyan", "Gold", "Coral", 
          "lime", "olive") 
  
# Wedge properties 
wp = { 'linewidth' : 3, 'edgecolor' : "red" } 
  
# Creating autocpt arguments 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n${:d}".format(pct, absolute) 
  
# Creating plot 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(Pub_InStateData,  
                                  autopct = lambda pct: func(pct, Pub_InStateData), 
                                  explode = BlowOut,  
                                  labels = States, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="black")) 
  
# Adding legend 
ax.legend(wedges, States, 
          title ="Breakdown", 
          loc ="center", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.rcParams['figure.facecolor'] = 'silver'
plt.setp(autotexts, size = 10, weight ="bold") 
ax.set_title("Public IN-State Tuition Breakdown") 
  
# show plot 
plt.show()
```
